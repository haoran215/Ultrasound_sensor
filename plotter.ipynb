The ADM process

```python
def ADM(s,thr,t_ref):
    up = []
    dn = []

    thr_UP = s[0]+thr
    thr_DN = s[0]-thr

    count_thr = 0
    for ti in range(1,len(s)):
        if(s[ti]>=thr_UP):
            if(count_thr>=t_ref):
                up.append(ti)
                count_thr = 0
                thr_DN = s[ti] - thr
                thr_UP = s[ti] + thr
            else:
                count_thr += 1
        elif(s[ti]<=thr_DN):
            if(count_thr>=t_ref):
                dn.append(ti)
                count_thr = 0
                thr_DN = s[ti] - thr
                thr_UP = s[ti] + thr
            else:
                count_thr +=1
    return up,dn
```

# ADM

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data and compute mean and Time_sec
file_path = r'testFiles/sensor_log_20250517_142352.csv'  # use forward slash or raw string
df = pd.read_csv(file_path)
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S')
df['Time_sec'] = (df['Time'] - df['Time'].iloc[0]).dt.total_seconds()
df['mean_signal'] = df[['Channel 1', 'Channel 2', 'Channel 3']].mean(axis=1)

# ADM algorithm
def ADM(s, thr, t_ref):
    up, dn = [], []
    thr_UP = s[0] + thr
    thr_DN = s[0] - thr
    count_thr = 0
    for ti in range(1, len(s)):
        if s[ti] >= thr_UP:
            if count_thr >= t_ref:
                up.append(ti)
                count_thr = 0
                thr_DN = s[ti] - thr
                thr_UP = s[ti] + thr
            else:
                count_thr += 1
        elif s[ti] <= thr_DN:
            if count_thr >= t_ref:
                dn.append(ti)
                count_thr = 0
                thr_DN = s[ti] - thr
                thr_UP = s[ti] + thr
            else:
                count_thr += 1
    return up, dn

# Apply ADM
up, dn = ADM(df['mean_signal'].to_numpy(), thr=100, t_ref=1)
time_up = df['Time_sec'].iloc[up]
time_dn = df['Time_sec'].iloc[dn]

# Determine spike marker positions
y_max = df['mean_signal'].max()
y_min = df['mean_signal'].min()
y_range = y_max - y_min
up_y = y_max + 0.05 * y_range   # slightly above the peak
dn_y = y_min - 0.05 * y_range   # slightly below the trough

# Plot signal with spikes above/below
plt.figure(figsize=(10, 5))
plt.plot(df['Time_sec'], df['mean_signal'], label='Mean Signal', color='steelblue')
plt.scatter(time_up, [up_y] * len(time_up), marker='|', label='UP spikes', color='red')
plt.scatter(time_dn, [dn_y] * len(time_dn), marker='|', label='DN spikes', color='green')

# Adjust y-limits so spikes are visible
plt.ylim(dn_y - 0.1 * y_range, up_y + 0.1 * y_range)

plt.xlabel('Time (s)')
plt.ylabel('Signal Amplitude')
plt.title('Ultrasound Signal with ADM Spikes (thr=100, t_ref=3)')
plt.legend(loc='center left')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 加载数据
file_path = 'Sensor02_s3_P.xlsx'
df = pd.read_excel(file_path)

# 提取压力与时间列
pressure = df['Corrected Pressure (hPa).8'].to_numpy()
time_sec = df['Time (s)'].to_numpy()

# 自动计算采样率
fs = 1 / np.mean(np.diff(time_sec))  # Hz
print(f"Estimated sampling rate: {fs:.2f} Hz")

# ADM 算法，同时检测 UP 和 DN 事件
def ADM(s, thr, t_ref_samples):
    up, dn = [], []
    thr_UP = s[0] + thr
    thr_DN = s[0] - thr
    count_thr = t_ref_samples + 1  # 保证第一个点能触发
    for ti in range(1, len(s)):
        if s[ti] >= thr_UP and count_thr >= t_ref_samples:
            up.append(ti)
            count_thr = 0
            thr_UP = s[ti] + thr
            thr_DN = s[ti] - thr
        elif s[ti] <= thr_DN and count_thr >= t_ref_samples:
            dn.append(ti)
            count_thr = 0
            thr_UP = s[ti] + thr
            thr_DN = s[ti] - thr
        else:
            count_thr += 1
    return up, dn

# 设置参数
thr = 0.5            # 阈值：0.02 hPa
t_ref_sec = 0.1       # 折返时间：0.1 秒
t_ref_samples = int(t_ref_sec * fs)

# 计算 UP/DN 事件索引
up_indices, dn_indices = ADM(pressure, thr, t_ref_samples)
time_up = time_sec[up_indices]
time_dn = time_sec[dn_indices]

# 绘图
plt.figure(figsize=(10,5))
plt.plot(time_sec, pressure, label='Right-Side Sensor Pressure', color='steelblue')

# UP 事件标记在曲线之上
y_max = pressure.max()
y_range = y_max - pressure.min()
up_y = y_max + 0.05 * y_range
plt.scatter(time_up, [up_y]*len(time_up), marker='|', color='red', label='UP spikes')

# DN 事件标记在曲线之下
dn_y = pressure.min() - 0.05 * y_range
plt.scatter(time_dn, [dn_y]*len(time_dn), marker='|', color='green', label='DN spikes')

# 格式设置
plt.ylim(dn_y - 0.1*y_range, up_y + 0.1*y_range)
plt.xlabel('Time (s)')
plt.ylabel('Pressure (hPa)')
plt.title(f'Right-Side Sensor Pressure with ADM Spikes (thr={thr}, t_ref={t_ref_sec}s)')
plt.legend(loc='lower left')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 加载数据
file_path = 'Sensor02_s3_P.xlsx'
df = pd.read_excel(file_path)

pressure = df['Corrected Pressure (hPa).8'].to_numpy()
time_sec = df['Time (s)'].to_numpy()

# 自动估计采样率
fs = 1 / np.mean(np.diff(time_sec))

def ADM_with_thr_trace(s, thr, t_ref_samples):
    up = []
    thr_UP_list = [s[0] + thr]
    thr_UP = thr_UP_list[-1]
    count_thr = t_ref_samples + 1
    for ti in range(1, len(s)):
        if s[ti] >= thr_UP and count_thr >= t_ref_samples:
            up.append(ti)
            count_thr = 0
            thr_UP = s[ti] + thr
        else:
            count_thr += 1
        thr_UP_list.append(thr_UP)
    return up, np.array(thr_UP_list)

# 参数设置
thr = 0.3
t_ref_sec = 0.1
t_ref_samples = int(t_ref_sec * fs)

# 获取 spike 和阈值轨迹
up_indices, thr_up_trace = ADM_with_thr_trace(pressure, thr, t_ref_samples)
time_up = time_sec[up_indices]

# 可视化
plt.figure(figsize=(12, 6))
plt.plot(time_sec, pressure, label='Right-Side Sensor Pressure', linewidth=1.5)
plt.plot(time_sec, thr_up_trace, '--', label='Dynamic Threshold (thr_UP)', color='orange', alpha=0.8)

# spike 标记
y_max = pressure.max()
y_range = y_max - pressure.min()
up_y = y_max + 0.05 * y_range
plt.scatter(time_up, [up_y] * len(time_up), marker='|', color='red', label='UP Spikes')

# spike 编号标注
for idx, ti in enumerate(up_indices):
    plt.text(time_sec[ti], up_y + 0.01 * y_range, f'{idx+1}', ha='center', fontsize=8, color='red')

# 图形美化
plt.ylim(min(pressure) - 0.1 * y_range, up_y + 0.1 * y_range)
plt.xlabel('Time (s)')
plt.ylabel('Pressure (hPa)')
plt.title('Right-Side Sensor Pressure with ADM UP Spikes and Dynamic Threshold')
plt.legend(loc='lower left')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 加载上传的 Excel 文件
file_path = 'Sensor02_s3_P.xlsx'
df = pd.read_excel(file_path)

# 提取感兴趣的列
pressure = df['Corrected Pressure (hPa).8'].to_numpy()
time_sec = df['Time (s)'].to_numpy()

# 计算采样率
fs = 1 / np.mean(np.diff(time_sec))

# 固定阈值 spike 检测函数
def FixedThresholdDetector(s, thr, t_ref_samples):
    up = []
    last_spike_idx = -np.inf
    for ti in range(1, len(s)):
        if s[ti] - s[ti - 1] > thr and ti - last_spike_idx > t_ref_samples:
            up.append(ti)
            last_spike_idx = ti
    return up

# 设置参数
thr = 0.1  # 阈值（单位：hPa）
t_ref_sec = 0.1
t_ref_samples = int(t_ref_sec * fs)

# 运行检测
up_indices = FixedThresholdDetector(pressure, thr, t_ref_samples)
time_up = time_sec[up_indices]

# spike 可视化准备
y_max = pressure.max()
y_range = y_max - pressure.min()
up_y = y_max + 0.05 * y_range

# 绘图
plt.figure(figsize=(12, 6))
plt.plot(time_sec, pressure, label='Right-Side Sensor Pressure', linewidth=1.5)
plt.scatter(time_up, [up_y] * len(time_up), marker='|', color='green', label='Threshold Spikes')

# 添加阈值和折返时间说明作为图例项
label_params = f'Threshold = {thr:.3f} hPa\nRefractory = {t_ref_sec:.2f} s'
plt.plot([], [], ' ', label=label_params)

plt.ylim(min(pressure) - 0.1 * y_range, up_y + 0.1 * y_range)
plt.xlabel('Time (s)')
plt.ylabel('Pressure (hPa)')
plt.title('Threshold Detection of Pressure Spikes')
plt.legend(loc='lower left')
plt.tight_layout()
plt.show()

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks, welch
from scipy import interpolate

# --- CONFIGURATION ---
FILENAME       = 'Sensor02_s2_P.xlsx'
TIME_COL       = 'Time (s)'
BASELINE_SEC   = 5.0         # seconds before onset to define noise window
SIGNAL_SEC     = 3.0         # seconds after onset to define signal window
THRESH_FRAC    = 0.1         # onset threshold = baseline + frac*(peak-baseline)
PSD_FS         = 1.0 / 0.1   # sampling frequency (Hz), = 1 / interval


# --- 1) Load data ---
df = pd.read_excel('Sensor02_s3_P.xlsx')
time = df[TIME_COL].to_numpy(dtype=float)

# assume all other columns are pressure trials
trial_cols = [c for c in df.columns if c != TIME_COL]
data = df[trial_cols].apply(pd.to_numeric, errors='coerce').to_numpy(dtype=float)

# --- 2) Helper: detect onset time for a single trace ---
def detect_onset(t, y, baseline_window=BASELINE_SEC, frac=THRESH_FRAC):
    # baseline = mean over first baseline_window seconds
    mask_base = t <= baseline_window
    baseline = np.mean(y[mask_base])
    peak     = np.max(y)
    thresh   = baseline + frac * (peak - baseline)
    # first crossing
    idx = np.argmax(y > thresh)
    return t[idx], idx, baseline

# --- 3) Loop over trials to compute SNR ---
snr_list = []
baseline_segments = []

for i, col in enumerate(trial_cols):
    y = data[:, i]
    t_on, idx_on, baseline0 = detect_onset(time, y)
    
    # define noise window: [t_on - BASELINE_SEC, t_on)
    mask_noise = (time >= t_on - BASELINE_SEC) & (time < t_on)
    noise_vals = y[mask_noise]
    
    # define signal window: [t_on, t_on + SIGNAL_SEC]
    mask_sig = (time >= t_on) & (time <= t_on + SIGNAL_SEC)
    sig_vals   = y[mask_sig]
    
    # compute metrics
    sigma_noise = np.nanstd(noise_vals)
    amp         = np.nanmax(sig_vals) - baseline0
    snr         = amp / sigma_noise if sigma_noise>0 else np.nan
    
    snr_list.append({
        'Trial': col,
        'Rise Time (s)': t_on,
        'Baseline Mean (mbar)': baseline0,
        'Noise σ (mbar)': sigma_noise,
        'Peak Amplitude (mbar)': amp,
        'SNR': snr
    })
    
    # store baseline for PSD
    baseline_segments.append(noise_vals)

df_snr = pd.DataFrame(snr_list)
mean_snr = df_snr['SNR'].mean()
sem_snr  = df_snr['SNR'].std(ddof=1) / np.sqrt(len(df_snr))

# --- 4) Bar-plot of SNRs ---
plt.figure(figsize=(8,4))
plt.bar(df_snr['Trial'], df_snr['SNR'], color='steelblue', alpha=0.7)
plt.axhline(mean_snr, color='orange', linestyle='--',
            label=f'Mean SNR = {mean_snr:.2f} ± {sem_snr:.2f} (SEM)')
plt.ylabel('SNR (unitless)')
plt.xlabel('Trial')
plt.title('Single‐Trial SNR')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# --- 5) Spectral analysis of the concatenated noise ---
# Concatenate all baseline segments (flatten)
noise_all = np.hstack(baseline_segments)

# Compute Welch PSD
f, Pxx = welch(noise_all, fs=PSD_FS, nperseg=256)

# Compute overall noise σ for reference
sigma_all = np.nanstd(noise_all)

plt.figure(figsize=(8,4))
plt.semilogy(f, np.sqrt(Pxx), label='Noise PSD')
plt.axhline(sigma_all, color='red', linestyle='--',
            label=f'Noise σ = {sigma_all:.3f} mbar')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Noise amplitude (mbar/√Hz)')
plt.title('Baseline Noise Power Spectral Density')
plt.legend()
plt.tight_layout()
plt.show()

# --- 6) Display SNR table ---
print(df_snr.to_string(index=False))
print(f"\nOverall Mean SNR = {mean_snr:.2f} ± {sem_snr:.2f} (SEM)")
```
